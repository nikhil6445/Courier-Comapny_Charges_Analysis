● Courier charges rate card at weight slab and pincode level. If the invoice mentions “Forward charges” then only forward charges (“fwd”) should be applicable as per zone and fixed & additional weights based on weight slabs. If the invoice mentions “Forward and rto charges” then forward charges (“fwd”) and RTO charges (“rto”) should be applicable as per zone and fixed & additional weights based on weight slabs.

● For the first 0.5 KG, “fixed” rate as per the slab is applicable. For each additional 0.5 KG, “additional” weight in the same proportion is applicable. Total charges will be “fixed” + “total additional” if any

```python
#Let's check Type of shipment we have in our dataframe

merge2_df['Type of Shipment'].unique()
```

```python
# Now let's make a function to get the expected charges as per X Company


def expected_chages_func(merge2_df):
    if (merge2_df["Type of Shipment"]=="Forward charges"):
        
        if (merge2_df["Delivery Zone as per X"]=='b'):
            
            if (merge2_df["Weight slab as per X (KG)"]<=0.5):
                return df_courier_rates["fwd_b_fixed"]
            else:
                return ((merge2_df["Weight slab as per X (KG)"]-0.5)*2*df_courier_rates["fwd_b_additional"])+df_courier_rates["fwd_b_fixed"]
                
             
        elif (merge2_df["Delivery Zone as per X"]=='d'):
            
            if (merge2_df["Weight slab as per X (KG)"]<=0.5):
                return df_courier_rates["fwd_d_fixed"]
            else:
                return ((merge2_df["Weight slab as per X (KG)"]-0.5)*2*df_courier_rates["fwd_d_additional"])+df_courier_rates["fwd_d_fixed"]   
        
        
        elif (merge2_df["Delivery Zone as per X"]=='e'):
            
            if (merge2_df["Weight slab as per X (KG)"]<=0.5):
                return df_courier_rates["fwd_e_fixed"]
            else:
                return ((merge2_df["Weight slab as per X (KG)"]-0.5)*2*df_courier_rates["fwd_e_additional"])+df_courier_rates["fwd_e_fixed"]   
            
    else:
        if (merge2_df["Delivery Zone as per X"]=='b'):
            
            if (merge2_df["Weight slab as per X (KG)"]<=0.5):
                return df_courier_rates["fwd_b_fixed"]+df_courier_rates["rto_b_fixed"]
            else:
                return ((merge2_df["Weight slab as per X (KG)"]-0.5)*2*(df_courier_rates["fwd_b_additional"]+df_courier_rates["rto_b_additional"]))+df_courier_rates["fwd_b_fixed"]+df_courier_rates["rto_b_fixed"]
        
        
        elif (merge2_df["Delivery Zone as per X"]=='d'):
            if (merge2_df["Weight slab as per X (KG)"]<=0.5):
                return df_courier_rates["fwd_d_fixed"]+df_courier_rates["rto_d_fixed"]
            else:
                return ((merge2_df["Weight slab as per X (KG)"]-0.5)*2*(df_courier_rates["fwd_d_additional"]+df_courier_rates["rto_d_additional"]))+df_courier_rates["fwd_d_fixed"]+df_courier_rates["rto_d_fixed"]

            
        elif (merge2_df["Delivery Zone as per X"]=='e'):
             if (merge2_df["Weight slab as per X (KG)"]<=0.5):
                return df_courier_rates["fwd_e_fixed"]+df_courier_rates["rto_e_fixed"]
             else:
                return ((merge2_df["Weight slab as per X (KG)"]-0.5))*2*df_courier_rates["fwd_e_additional"]+df_courier_rates["rto_e_additional"]+df_courier_rates['fwd_e_fixed']+df_courier_rates['rto_e_fixed']
```

- Two types of shipment we can see here

```python
# let's find out the expected charge using apply method
merge2_df['Expected Charge as per X (Rs.)'] = merge2_df.apply(expected_chages_func,axis=1)
```
